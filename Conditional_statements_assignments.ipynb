```python
# Exercise 4: Calculator Discount A store offers:
# 10% discount if purchase > $100
# 5% discount if purchase > $50
# No discount otherwise
# Calculate the final price.
print("\nExercise 4: Calculator Discount A store offers:")
print("10% discount if purchase > $100")
print("5% discount if purchase > $50")
print("No discount otherwise")

org_item = float(input("Enter amout"))
item1 = org_item 
if item1 > 100:
    discount = 0.1
    item1 = item1 * (1-discount)
elif item1 > 50:
    discount = 0.05
    item1 = item1 *(1-discount)
print("Final Price :",item1)
```

```python
# Exercise 5: Leap Year Checker Write a program to check if a year is a leap year:
# Divisible by 4 AND (not divisible by 100 OR divisible by 400)
print("\nExercise 5: Leap Year Checker Write a program to check if a year is a leap year:")
print("\nDivisible by 4 AND (not divisible by 100 OR divisible by 400)")
year1 = int(input("Enter Year"))
if (year1%4 == 0 and year1%100 != 0) or (year1%400 ==0):
    print(f"{year1} is leap year")
else:
    print(f"{year1} is not leap year")
```

```python
# Exercise 6: Grade Calculator with Bonus Calculate grades (A-F)
# but add +1 bonus grade level if attendance > 90%.
print("\nExercise 6: Grade Calculator with Bonus Calculate grades (A-F)")
print("\nadd +1 bonus grade level if attendance > 90%.")
grades = ["A","B","C","D","E","F"]
attendence = int(input("Enter Attendence percentage"))
current_grade = input("Enter Present Grade").upper()

if current_grade in grades:
    grade_index = grades.index(current_grade)
    if attendence > 90 and grade_index >0 :
        grade_index -= 1
    final_grade = grades[grade_index]
    print(f"Final Grade : {final_grade}")        
else:    
    print("Invalid Grade")
```

```python
# Exercise 7: Login System with Attempts Create a login system that:
# Checks username and password
# Tracks login attempts (max 3)
# Locks account after failed attempts

# Method 1

print("\nExercise 7: Login System with Attempts Create a login system that:")
print("\nChecks username and password")
print("Tracks login attempts (max 3)")
print("Locks account after failed attempts")

users = {'aditya' : 'aditya234'}

max_attempts = 3
attempts =0
while attempts < max_attempts:
    username = input("Enter username")
    password = input("Enter password")

    if username in users and users[username] == password:
        print("Login successfully")
        break
    else:
        attempts +=1 
        print("login failed ")
        print("no of attempts" , max_attempts-attempts)
if attempts == max_attempts:
    print("Account locked")
```
