# 1. Even or Odd

## Logic
- Take a number as input.
- Divide the number by 2.
- Check the remainder using modulus operator `%`.
- If remainder is 0 → number is Even.
- Otherwise → number is Odd.

```python
def number_check():
    x = int(input())
    if x %2 ==0:
        print("Even number")
    else:
        print("odd Number")
```

```python
number_check()
```

# 2. Factorial of a Number

## Logic
- Take a number `n`.
- Start with a variable `result = 1`.
- Multiply `result` by all numbers from 1 to `n`.
- Return the final value of `result`.
- Special case: factorial of 0 is 1.

**Example:**
5! = 1 × 2 × 3 × 4 × 5

```python
total =1 
def factorial():
    global total
    n = int(input())
    print(n)
    while n !=0:
        total *= n
        n = n-1
    print(total)
```

```python
factorial()
```

```python
def factorial(n):
    total = 1
    while n != 0:
        total *= n
        n -= 1
    return total

num = int(input())
print(num)
print(factorial(num))
```

# 3. Reverse a String

## Logic

### Method 1 (Loop)
- Take a string.
- Start from the last character.
- Add each character to a new string.
- Return the new reversed string.

### Method 2
- Use slicing with negative step.

```python
# method 1
def rev_string1():
    a = input()
    result = ""
    for letter in range(len(a)-1,-1,-1):
        result +=a[letter]

    return result
```

```python
rev_string1()
```

```python
#Method 2 slicing
def rev_string2():
    a = input()
    result = a[::-1]
    return result
```

```python
rev_string2()
```

# 4. Check Prime Number

## Logic
- Prime numbers are greater than 1.
- Check if number is less than or equal to 1 → Not prime.
- Try dividing the number from 2 to `n-1`.
- If divisible by any number → Not prime.
- If not divisible by any → Prime.

```python
def prime_num():
    a = int(input())
    print(f"NUmber is {a}")
    if a <= 1:
        print("Not Prime")
        return

    for i in range(2, int(a**0.5) + 1):
        if a % i == 0:
            print("Not Prime")
            return

    print("Prime")
```

```python
prime_num()
```

# 5. Largest Number in List

## Logic
- Assume first element is largest.
- Compare it with each element in the list.
- If a bigger number is found → update largest.
- Return the largest value.

```python
def largest_num(numbers):
    largest = numbers[0]
    for num in numbers:
        if num > largest:
            largest = num 
        
    return largest
```

```python
num1=[1,2,3,4,5,6,7,8,5,2,23,54,235,0]
largest_num(num1)
```

# 7. Count Vowels

## Logic
- Create a string of vowels (`aeiouAEIOU`).
- Set `count = 0`.
- Loop through each character.
- If character is in vowels → increase `count`.
- Return `count`.


```python
def count_vowels(s):
    vowels = "aeiou"
    count = 0
    for i in s.lower():
        if i in vowels:
            print(i)
            count += 1
    return count
```

# 8. Fibonacci Sequence

## Logic
- First two numbers are 0 and 1.
- Next number = sum of previous two.
- Repeat until required `n` terms.
- Store numbers in a list.
- Return the list.
**Example:**
0, 1, 1, 2, 3, 5, 8…

```python
def fibonacci_series(n):
    a, b = 0, 1
    result = []

    for _ in range(n):
        result.append(a)
        a, b = b, a + b

    return result
```

```python
fibonacci_series(1)
```

# 9. Palindrome Check

## Logic
- Take a string.
- Reverse the string.
- Compare original string with reversed string.
- If same → Palindrome.
- Else → Not palindrome.

**Example:**
"madam" → same when reversed


```python
def palindrome(str1):
    str2 = str1[::-1]
    if str1 == str2:
        print("YES")
    else:
        print("no")
```

```python
def palindrome(s):
    s = s.lower().replace(" ", "")
    return s == s[::-1]
```

```python
palindrome("madam")
```

# 10. GCD (Greatest Common Divisor)

## Logic (Euclidean Method)
- Take two numbers `a` and `b`.
- While `b` is not 0:
  - Replace `a` with `b`.
  - Replace `b` with remainder of `a % b`.
- When `b` becomes 0 → `a` is GCD.

```python
def GCD(a, b):
    while b != 0:
        a, b = b, a % b   # correct step
    return a
```

```python
GCD(2,4)
```

# 11. Remove Duplicates from List

## Logic
- Create empty list.
- Loop through original list.
- If element is not already in new list → add it.
- Return new list.

```python
def remove_Duplicates(list1):
    list2 = []
    for num in list1:
        if num not in list2:
            list2.append(num)
        
    return list2
```

# 12. Get Even Numbers from List

## Logic
- Create empty list.
- Loop through original list.
- If number `% 2 == 0` → add to new list.
- Return new list.

```python
def even_num(list1):
    list2 = []
    for num in list1:
        if num %2==0:
            list2.append(num)

    return list2
```

```python
def even_num(list1):
    return [num for num in list1 if num % 2 == 0]
```

```python
list1 = range(1,20)
even_num(list1)
```

# 13. Character Frequency in String

## Logic
- Create empty dictionary.
- Loop through each character.
- If character already in dictionary:
  - Increase its count.
- Else:
  - Add character with count 1.
- Return dictionary.


```python
def freq(s):
    dict1 = {}
    for ch in s:
        if ch in dict1:
            dict1[ch] += 1
        else:
            dict1[ch] = 1
    return dict1
```

```python
freq("Lakshman Aditya")
```

# 14. Swap Two Numbers (Without Third Variable)

## Logic

### Method 1 (Python way)
`a, b = b, a`

### Method 2 (Math way)
- `a = a + b`
- `b = a - b`
- `a = a - b`

```python
def swap(a,b):
    a = a+b
    b=a-b
    a=a-b
    return a,b
    
```

```python
swap(1,2)
```
