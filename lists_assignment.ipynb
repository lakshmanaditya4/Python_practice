```python
print("\nQuestion 34: Count the depth of nesting in [1, [2, 3], [4, [5, 6]], 7]")

list1 = [1, [2, 3], [4, [5, 6]], 7]
depth = 1
for i in list1:
    if type(i) == list:
        depth += 1
        for j in i:
            if type(j) == list: 
                depth += 1
                for k in j:
                    if type(k) == list:
                        depth +=1
                        break

print("Depth is", depth)
```
