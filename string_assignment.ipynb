```python
# Question 42: Convert "hello world" to Morse code
print("\nQuestion 42: Convert 'hello world' to Morse code")

morse_code = {
    'a': '.-',    'b': '-...',  'c': '-.-.',  'd': '-..',
    'e': '.',     'f': '..-.',  'g': '--.',   'h': '....',
    'i': '..',    'j': '.---',  'k': '-.-',   'l': '.-..',
    'm': '--',    'n': '-.',    'o': '---',   'p': '.--.',
    'q': '--.-',  'r': '.-.',   's': '...',   't': '-',
    'u': '..-',   'v': '...-',  'w': '.--',   'x': '-..-',
    'y': '-.--',  'z': '--..',
    ' ': '/'
}

text = "hello world"
result = ""

for ch in text:
    result += morse_code[ch] + " "

print(result.strip())
```

```python
# Question 43: Find the longest common substring between "programming" and "grammar"
# method 1
print("\nQuestion 43: Find the longest common substring between 'programming' and 'grammar'")
str1 = "programming"
str2 = "grammer"
longest = ""
for i in range(len(str1)):
    for j in range(i+1 ,len(str1)+1):
        sub = str1[i:j]
        if sub in str2 and len(sub)>len(longest):
            longest = sub
print(longest)
```

```python
# Question 43: Find the longest common substring between "programming" and "grammar"
# method 2
print("\nQuestion 43: Find the longest common substring between 'programming' and 'grammar'")
str1 = "programming"
str2 = "grammar"
max_len = 0
result = ""
for i in range(len(str1)):
    for j in range(len(str2)):
        temp = ""
        x, y = i, j         #here x,y are pointer to move forward in str1 and str2 
        while x < len(str1) and y < len(str2) and str1[x] == str2[y]:
            temp += str1[x]
            x += 1
            y += 1

        if len(temp) > max_len:
            max_len = len(temp)
            result = temp

print("Longest common substring:", result)
```

```python
# Question 49: Convert "hello world" to ROT13 encoding
print("\nQuestion 49: Convert 'hello world' to ROT13 encoding")
s = "hello world"
result = ""

for ch in s:
    if ch >= 'a' and ch <= 'z':
        new_char = chr((ord(ch) - ord('a') + 13) % 26 + ord('a'))
        result += new_char
    elif ch >= 'A' and ch <= 'Z':
        new_char = chr((ord(ch) - ord('A') + 13) % 26 + ord('A'))
        result += new_char
    else:
        result += ch

print(result)
```
