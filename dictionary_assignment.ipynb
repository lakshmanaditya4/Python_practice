```python
# Question 9: Count frequency of each character in string "hello"
print("\nQuestion 9: Count frequency of each character in string 'hello'")
str1 = "hello"
freq = {}
for ch in str1:
    if ch in freq:
        freq[ch]+= 1
    else:
        freq[ch] = 1
print(freq)
```
