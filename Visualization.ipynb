```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

data = {
    "Employee_ID": range(1, 101),
    "Age": np.random.randint(21, 60, 100),
    "Salary": np.random.randint(20000, 120000, 100),
    "Experience": np.random.randint(1, 20, 100),
    "Department": np.random.choice(["IT", "HR", "Finance", "Sales"], 100),
    "Performance_Score": np.random.randint(1, 5, 100)
}

df = pd.DataFrame(data)
print(df.head())
```

```python
df
```

```python
# 1. Line Plot
plt.figure(figsize = (10,5))
plt.plot(df["Employee_ID"], df["Salary"],color='#006000') # x-axis = emp_id and y - axis is Salary
plt.xlabel("Emplpoyee id")              # for color you can search "hexa codes for colors "
plt.ylabel("Salary")
plt.title("Salary Trend")
plt.show()
```

```python
# 1. Line Plot               marker = 'o' for dot also marker ranges from 0 to 11 a
plt.figure(figsize=(10,5))
plt.plot(df["Employee_ID"], df["Salary"],marker= 1) 
plt.xlabel("Emplpoyee id")           
plt.ylabel("Salary")
plt.title("Salary Trend")
plt.show()
```

```python
# 1. Line Plot
plt.figure(figsize=(10,5))
plt.plot(df["Employee_ID"], df["Salary"],color='Blue',marker='o',markerfacecolor='r',
         markeredgecolor='b')
plt.xlabel("Emplpoyee id")
plt.ylabel("Salary")
plt.title("Salary Trend")
plt.show()
```

```python
# 1. Line Plot
plt.figure(figsize=(20,5))
plt.plot(df["Employee_ID"], df["Salary"],linewidth=4,markersize=16,color='Blue',marker='o',markerfacecolor='w',
         markeredgecolor='b')
plt.xlabel("Emplpoyee id")
plt.ylabel("Salary")
plt.title("Salary Trend")
plt.show()
```

```python
# 1. Line Plot                      alpha forlight 
plt.figure(figsize=(20,5))
plt.plot(df["Employee_ID"], df["Salary"],linewidth=10,alpha=0.1,markersize=25,color='Blue',marker='o',markerfacecolor='w',
         markeredgecolor='b')
plt.xlabel("Emplpoyee id")
plt.ylabel("Salary")
plt.title("Salary Trend")
plt.show()
```
